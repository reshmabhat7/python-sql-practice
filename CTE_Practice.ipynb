```python
import sqlite3
import pandas as pd

# Connect to a new in-memory database (or use a .db file if you want to save)
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
```

```python
# Sample table with mock data
cursor.execute("""
CREATE TABLE orders (
    Order_ID INTEGER,
    Customer_ID TEXT,
    Region TEXT,
    Product_Name TEXT,
    Sales REAL,
    Profit REAL
)
""")

# Insert mock data
orders_data = [
    (1, 'C001', 'East', 'Chair', 250, 50),
    (2, 'C002', 'West', 'Table', 400, 100),
    (3, 'C001', 'East', 'Desk', 300, 40),
    (4, 'C003', 'South', 'Chair', 150, 20),
    (5, 'C001', 'East', 'Lamp', 120, 30),
    (6, 'C002', 'West', 'Chair', 200, 60),
    (7, 'C004', 'North', 'Desk', 500, 150),
    (8, 'C003', 'South', 'Lamp', 180, 25),
    (9, 'C002', 'West', 'Desk', 300, 70),
]
cursor.executemany("INSERT INTO orders VALUES (?, ?, ?, ?, ?, ?)", orders_data)
conn.commit()
```

```python
query1 = """
WITH avg_sales AS (
    SELECT Region, AVG(Sales) AS avg_sales
    FROM orders
    GROUP BY Region
)
SELECT * FROM avg_sales;
"""
pd.read_sql(query1, conn)
```

```python
query2 = """
WITH order_counts AS (
    SELECT Customer_ID, COUNT(Order_ID) AS total_orders
    FROM orders
    GROUP BY Customer_ID
)
SELECT * FROM order_counts
WHERE total_orders > 2;
"""
pd.read_sql(query2, conn)
```

```python
query3 = """
WITH product_profits AS (
    SELECT Product_Name, SUM(Profit) AS total_profit
    FROM orders
    GROUP BY Product_Name
)
SELECT * FROM product_profits
ORDER BY total_profit DESC
LIMIT 5;
"""
pd.read_sql(query3, conn)
```
